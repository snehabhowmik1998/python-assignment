### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Step 2. Import the dataset from the AutoMPG dataset file. Assign it to a variable called mpg_df

```python
df=pd.read_csv("https://raw.githubusercontent.com/joanby/python-ml-course/master/datasets/auto/auto-mpg.csv")
```

### Step 4. Check and handle the missing values, if any.

```python
df.isnull().sum() 
```

```python
df["mpg"]=df["mpg"].fillna(df["mpg"].mean()) # replacing the null values with the mean
```

```python
df["horsepower"]=df["horsepower"].fillna(df["horsepower"].mean())
```

```python
df.isnull().sum()
```

### Step 5. Create a plot to check the relationship between horsepower and acceleration. Note down your insight for the same. Beautify the graph using various customizations.

```python
sns.relplot(x="horsepower",y="acceleration",data=df,color="blue",marker="o")
plt.title("relationship between horsepower and acceleration",loc="Right",color="red")
plt.xlabel("horsepower",color="orange")
plt.ylabel("acceleration",color="green")
plt.show()
```

### Step 6. Generate subplots to display the histograms for acceleration, displacement and weight and kilometer_per_litre. Note down your insight for the same

```python
plt.subplot(2,1,1)
plt.scatter(df["acceleration"],df["displacement"])
plt.title("histoigram of acceleration & displacement")
plt.xlabel("acceleration")
plt.ylabel("displacement")
plt.show()

plt.subplot(2,1,2)
plt.scatter(df["weight"],df["mpg"])
plt.title("histoigram of weight & mpg")
plt.xlabel("weight")
plt.ylabel("mpg")
plt.show()
```

### Step 7. Create a plot to check the relationship between displacement and weight for origin 2, color the datapoints based on no of cylinders. Note down your insight for the same and save the plot as an image file. 

```python
origin_df=df[df["origin"]==2] # here i am making a new data set in which i am changing the data set in which origin is 2
origin_df
```

```python
sns.relplot(x=df["displacement"],y=df["weight"],hue="cylinders",data=origin_df,palette='Set2')
xlabel=("displacement")
ylabel=("weight")
plt.show()
```

### Step 9. Check the relationship of multiple variables wrt kilometer_per_litre. Note down your insight for the same

### Step 10. Display the average weight based on no of cylinders present. Note down the insight for the same

```python
avg_weight_of_cylinders=df.groupby("cylinders").weight.mean() # using groupby function 
avg_weight_of_cylinders
```
